```python
# format Y2021_df
# convert cycle date to datetime
Y2021_df['Cycle_Date'] = pd.to_datetime(Y2021_df['Cycle_Date'])
Y2021_df['Cycle_Date'] = Y2021_df['Cycle_Date'].dt.strftime('%m-%d-%Y')

# Format the columns have the comma for a thousands separator.
Y2021_df["Total_Assets"] = Y2021_df["Total_Assets"].map("${:,}".format)
Y2021_df["Current_mbrs"] = Y2021_df["Current_mbrs"].map("{:,}".format)
Y2021_df["#New_Auto"] = Y2021_df["#New_Auto"].map("{:,}".format)
Y2021_df["$New_Auto"] = Y2021_df["$New_Auto"].map("${:,}".format)
Y2021_df["#Used_Auto"] = Y2021_df["#Used_Auto"].map("{:,}".format)
Y2021_df["$Used_Auto"] = Y2021_df["$Used_Auto"].map("${:,}".format)
Y2021_df["#Lease"] = Y2021_df["#Lease"].map("{:,}".format)
Y2021_df["$Lease"] = Y2021_df["$Lease"].map("${:,}".format)
Y2021_df["#Other_Secured"] = Y2021_df["#Other_Secured"].map("{:,}".format)
Y2021_df["$Other_Secured"] = Y2021_df["$Other_Secured"].map("${:,}".format)
Y2021_df["#CC"] = Y2021_df["#CC"].map("{:,}".format)
Y2021_df["$CC"] = Y2021_df["$CC"].map("${:,}".format)
Y2021_df["#Other_Unsecured"] = Y2021_df["#Other_Unsecured"].map("{:,}".format)
Y2021_df["$Other_Unsecured"] = Y2021_df["$Other_Unsecured"].map("${:,}".format)
Y2021_df["#1st_Mtg"] = Y2021_df["#1st_Mtg"].map("{:,}".format)
Y2021_df["$1st_Mtg"] = Y2021_df["$1st_Mtg"].map("${:,}".format)
Y2021_df["#2nd_Mtg"] = Y2021_df["#2nd_Mtg"].map("{:,}".format)
Y2021_df["$2nd_Mtg"] = Y2021_df["$2nd_Mtg"].map("${:,}".format)

# format net worth ratio
Y2021_df['Net_worth_ratio'] = Y2021_df['Net_worth_ratio'].astype(float)
Y2021_df['Net_worth_ratio'] = Y2021_df['Net_worth_ratio'] / 100

# format rate columns
Y2021_df['New_Auto_Rate'] = Y2021_df['New_Auto_Rate'] / 100
Y2021_df['Used_Auto_Rate'] = Y2021_df['Used_Auto_Rate'] / 100
Y2021_df['Lease_Rate'] = Y2021_df['Lease_Rate'] / 100
Y2021_df['Other_Secured_Rate'] = Y2021_df['Other_Secured_Rate'] / 100
Y2021_df['CC_Rate'] = Y2021_df['CC_Rate'] / 100
Y2021_df['Other_Unsecured_Rate'] = Y2021_df['Other_Unsecured_Rate'] / 100
Y2021_df['1st_Mtg_Rate'] = Y2021_df['1st_Mtg_Rate'] / 100
Y2021_df['2nd_Mtg_Rate'] = Y2021_df['2nd_Mtg_Rate'] / 100

Y2021_df
```
