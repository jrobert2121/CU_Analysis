```python
import pandas as pd
```

```python
# read in text files into dataframe
foicu_df = pd.read_csv('2022_Q1/FOICU.txt', sep = ',')
fs220_df = pd.read_csv("2022_Q1/FS220.txt", sep = ',')
fs220a_df = pd.read_csv('2022_Q1/FS220A.txt', sep = ',')
fs220b_df = pd.read_csv('2022_Q1/FS220B.txt', sep = ',')
fs220d_df = pd.read_csv('2022_Q1/FS220D.txt', sep = ',')
fs220g_df = pd.read_csv('2022_Q1/FS220G.txt', sep = ',')
fs220l_df = pd.read_csv('2022_Q1/FS220L.txt', sep = ',')
fs220m_df = pd.read_csv('2022_Q1/FS220M.txt', sep = ',')
fs220p_df = pd.read_csv('2022_Q1/FS220P.txt', sep = ',')
```

```python
# create Q1_2022_df
Q1_2022_data = [foicu_df['CU_NUMBER'], foicu_df['CYCLE_DATE'], foicu_df['CU_NAME'], foicu_df['STATE'],
        fs220_df['ACCT_010'], fs220_df['ACCT_083'], fs220d_df['Acct_700'], fs220a_df['Acct_998'], fs220_df['ACCT_523'],
               fs220b_df['ACCT_958'], fs220a_df['ACCT_385'], fs220_df['ACCT_524'], fs220_df['ACCT_968'], fs220a_df['ACCT_370'],
               fs220a_df['ACCT_565'], fs220a_df['Acct_954'], fs220a_df['ACCT_002'], fs220l_df['ACCT_595B'],
               fs220m_df['ACCT_963C'], fs220l_df['ACCT_698C'], fs220_df['ACCT_521'], fs220b_df['ACCT_993'],
               fs220a_df['ACCT_396'], fs220_df['ACCT_522'], fs220b_df['ACCT_994'], fs220a_df['ACCT_397'],
               fs220l_df['ACCT_563A'], fs220m_df['ACCT_959A'], fs220l_df['ACCT_703A'], fs220l_df['ACCT_562A'],
               fs220m_df['ACCT_960A'], fs220l_df['ACCT_386A']]

# create headers
Q1_2022_headers = ['CU_Number', 'Cycle_Date', 'CU_Name', 'State', 'Total_Assets', 'Current_mbrs',
          'Classification', 'Net_worth_ratio', 'New_Auto_Rate', '#New_Auto', '$New_Auto', 'Used_Auto_Rate', '#Used_Auto',
                  '$Used_Auto', 'Lease_Rate', '#Lease', '$Lease', 'Other_Secured_Rate', '#Other_Secured', '$Other_Secured',
                  'CC_Rate', '#CC', '$CC', 'Other_Unsecured_Rate', '#Other_Unsecured', '$Other_Unsecured',
                  '1st_Mtg_Rate', '#1st_Mtg', '$1st_Mtg', '2nd_Mtg_Rate', '#2nd_Mtg', '$2nd_Mtg']

Q1_2022_df = pd.concat(Q1_2022_data, axis=1, keys=Q1_2022_headers)
```

```python
# format Q1_2022_df
# convert cycle date to datetime
Q1_2022_df['Cycle_Date'] = pd.to_datetime(Q1_2022_df['Cycle_Date'])
Q1_2022_df['Cycle_Date'] = Q1_2022_df['Cycle_Date'].dt.strftime('%m-%d-%Y')

# Format the columns have the comma for a thousands separator.
Q1_2022_df["Total_Assets"] = Q1_2022_df["Total_Assets"].map("${:,}".format)
Q1_2022_df["Current_mbrs"] = Q1_2022_df["Current_mbrs"].map("{:,}".format)
Q1_2022_df["#New_Auto"] = Q1_2022_df["#New_Auto"].map("{:,}".format)
Q1_2022_df["$New_Auto"] = Q1_2022_df["$New_Auto"].map("${:,}".format)
Q1_2022_df["#Used_Auto"] = Q1_2022_df["#Used_Auto"].map("{:,}".format)
Q1_2022_df["$Used_Auto"] = Q1_2022_df["$Used_Auto"].map("${:,}".format)
Q1_2022_df["#Lease"] = Q1_2022_df["#Lease"].map("{:,}".format)
Q1_2022_df["$Lease"] = Q1_2022_df["$Lease"].map("${:,}".format)
Q1_2022_df["#Other_Secured"] = Q1_2022_df["#Other_Secured"].map("{:,}".format)
Q1_2022_df["$Other_Secured"] = Q1_2022_df["$Other_Secured"].map("${:,}".format)
Q1_2022_df["#CC"] = Q1_2022_df["#CC"].map("{:,}".format)
Q1_2022_df["$CC"] = Q1_2022_df["$CC"].map("${:,}".format)
Q1_2022_df["#Other_Unsecured"] = Q1_2022_df["#Other_Unsecured"].map("{:,}".format)
Q1_2022_df["$Other_Unsecured"] = Q1_2022_df["$Other_Unsecured"].map("${:,}".format)
Q1_2022_df["#1st_Mtg"] = Q1_2022_df["#1st_Mtg"].map("{:,}".format)
Q1_2022_df["$1st_Mtg"] = Q1_2022_df["$1st_Mtg"].map("${:,}".format)
Q1_2022_df["#2nd_Mtg"] = Q1_2022_df["#2nd_Mtg"].map("{:,}".format)
Q1_2022_df["$2nd_Mtg"] = Q1_2022_df["$2nd_Mtg"].map("${:,}".format)

# format net worth ratio
Q1_2022_df['Net_worth_ratio'] = Q1_2022_df['Net_worth_ratio'].astype(float)
Q1_2022_df['Net_worth_ratio'] = Q1_2022_df['Net_worth_ratio'] / 100

# format rate columns
Q1_2022_df['New_Auto_Rate'] = Q1_2022_df['New_Auto_Rate'] / 100
Q1_2022_df['Used_Auto_Rate'] = Q1_2022_df['Used_Auto_Rate'] / 100
Q1_2022_df['Lease_Rate'] = Q1_2022_df['Lease_Rate'] / 100
Q1_2022_df['Other_Secured_Rate'] = Q1_2022_df['Other_Secured_Rate'] / 100
Q1_2022_df['CC_Rate'] = Q1_2022_df['CC_Rate'] / 100
Q1_2022_df['Other_Unsecured_Rate'] = Q1_2022_df['Other_Unsecured_Rate'] / 100
Q1_2022_df['1st_Mtg_Rate'] = Q1_2022_df['1st_Mtg_Rate'] / 100
Q1_2022_df['2nd_Mtg_Rate'] = Q1_2022_df['2nd_Mtg_Rate'] / 100

Q1_2022_df
```

```python
# change index to CU_Number
Q1_2022_df.set_index('CU_Number')
```

```python
# write df to excel
Q1_2022_df.to_excel('MSUFCU.xlsx', sheet_name='Q1_2022')
```
